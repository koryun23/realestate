```python
import requests
from bs4 import BeautifulSoup
import pandas
r = requests.get("https://www.sports.ru/")
c = r.content
soup = BeautifulSoup(c, "html.parser")

all_events = soup.find_all("div", {"class":"accordion-group"})
l = []
for event in all_events:
    d = {}
    all_games = []
    if "хоккей" in event.find("div", {"class":"accordion__handle"}).find("span", {"class":"accordion__title"}).find("a").text:
        tournament= event.find("div", {"class":"accordion__handle"}).find("span", {"class":"accordion__title"}).find("a").text
        main_event = event.find("ul", {"class": "teaser"})
        all_games = main_event.find_all("li", {"class": "teaser-event"})
    for game in all_games:
        status = game.find("div", {"class": "teaser-event__status"}).find("span").text
        first_team = game.find("div", {"class": "teaser-event__board"}).find_all("div", {"class": "teaser-event__board-player"})[0].find("a").text
        second_team = game.find("div", {"class": "teaser-event__board"}).find_all("div", {"class": "teaser-event__board-player"})[1].find("a").text
        score1 = game.find("div", {"class": "teaser-event__board"}).find("a", {"class": "teaser-event__board-score"}).find_all("span")[0].text
        score2 = game.find("div", {"class": "teaser-event__board"}).find("a", {"class": "teaser-event__board-score"}).find_all("span")[1].text
        
        score = score1 + "-" + score2
        
        output = status + "    " + first_team + "  " + score + "  " + second_team
        d['Tournament'] = tournament
        d['Status'] = status
        d['Team1'] = first_team
        d['Score'] = score
        d['Team2'] = second_team
        l.append({"Tournament": tournament, "Status": status, "Team1": first_team, "Score": score, "Team2": second_team})
df = pandas.DataFrame(l)
df
```
